##### 2. Создайте функцию, которая принимает как аргументы целое число N и первый элемент (вещественное число), и разность (вещественное число) и создает матрицу numpy по диагонали, которой располагаются первые N членов арифметической прогрессии.

```python
def matrix(N, a, d):
    el = np.arange(a, a + N * d, d)
    diag_matrix = np.diag(el)
    return diag_matrix

# Пример использования функции
N = 5
a = 2.0
d = 3.0

result_matrix = matrix(N, a, d)
print(result_matrix)
```

##### 5. Создайте две матрицы размером (5,5). Одна матрица содержит 5 в шахматном порядке как в задаче домашнего задания, другая имеет треугольную форму содержащую 5 на основной диагонали и в позициях выше ее, а ниже все 0. Посчитайте их детерминант и найдите обратные матрицы.

```python
# Создаем матрицу в шахматном порядке
chessboard_matrix = np.tile([[1, 0], [0, 1]], (2, 2))

# Создаем треугольную матрицу
triangle_matrix = np.triu(np.ones((5, 5)))

# Выводим обе матрицы
print("Матрица в шахматном порядке:")
print(chessboard_matrix)
print("\nТреугольная матрица:")
print(triangle_matrix)

# Вычисляем детерминанты
det_chessboard = np.linalg.det(chessboard_matrix)
det_triangle = np.linalg.det(triangle_matrix)

print("\nДетерминант матрицы в шахматном порядке:", det_chessboard)
print("Детерминант треугольной матрицы:", det_triangle)

# Находим обратную матрицу только для треугольной матрицы
inv_triangle = np.linalg.inv(triangle_matrix)

print("\nОбратная матрица для треугольной матрицы:")
print(inv_triangle)
```

##### 11. Выделите в отдельную таблицу последние десять строк. Уберите в ней столбец с ценой. Склейте ее с первоначальной таблицей при помощи append. Заполните отсутствующие значения цены средним по таблице.

```python
# Выделение последних десяти строк в отдельную таблицу и удаление столбца с ценой
last_rows = dat.tail(10).drop(columns=['Цена'])

# Склеивание таблиц с помощью метода concat
dat_with_last_rows = pd.concat([dat, last_rows])

# Заполнение отсутствующих значений цены средним по таблице
mean_price = dat['Цена'].mean()
dat_with_last_rows['Цена'] = dat_with_last_rows['Цена'].fillna(mean_price)

# Вывод обновленной таблицы
print(dat_with_last_rows)
```

##### 12. Выделите пять последних колонок в отдельную таблицу. Удалите в ней строки, в которых цена ниже среднего. Присоедините эту таблицу к изначальной (выберите самый подходящий тип соединения).

```python
# Выделение пяти последних колонок в отдельную таблицу
last_columns = dat.iloc[:, -5:]

# Удаление строк, в которых цена ниже среднего
mean_price = dat['Цена'].mean()
last_columns = last_columns[dat['Цена'] >= mean_price]

# Присоединение таблицы к изначальной с помощью метода merge
merged_table = dat.merge(last_columns, left_index=True, right_index=True)

# Вывод обновленной таблицы
print(merged_table)
```

##### 13. Выведите таблицу, содержащую среднюю цену и количество квартир на каждом этаже из первоначального набора данных.

```python
# Группировка данных по этажам и вычисление средней цены и количества квартир на каждом этаже
floor_stats = dat.groupby('Этаж').agg({'Цена': 'mean', 'Количество комнат': 'count'})

# Переименование столбцов для ясности
floor_stats = floor_stats.rename(columns={'Цена': 'Средняя цена', 'Количество комнат': 'Количество квартир'})

# Вывод таблицы с результатами
print(floor_stats)
```
