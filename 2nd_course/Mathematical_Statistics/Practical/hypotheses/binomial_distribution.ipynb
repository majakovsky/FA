```python
import numpy as np
from scipy.stats import binom
from scipy.stats import chisquare

# Шаг 2: Задайте значения параметров биноминального распределения
n = 10  # количество испытаний
p = 0.5  # вероятность успеха

# Шаг 3: Сгенерируйте случайную выборку Х размером N из биноминального распределения
N = 1000
sample = np.random.binomial(n, p, N)

# Шаг 4: Посчитайте эмпирические частоты для каждого значения успеха в выборке Х.
ef = np.bincount(sample)

# Шаг 5: Используя формулу Бернулли, вычислите теоретические частоты
tf = binom.pmf(np.arange(n + 1), n, p) * N

# Шаг 6: Сравните теоретические и эмпирические частоты
print("Эмпирические частоты:", ef)
print("Теоретические частоты:", tf)

# Шаг 7: Проведите статистический тест (тест хи-квадрат) для проверки гипотезы
chi2_stat, p_value = chisquare(f_obs=ef, f_exp=tf)

# Шаг 8: Выведите результаты теста
print("\nСтатистическая статистика (chi2):", chi2_stat)
print("p-значение:", p_value)

# Шаг 9: Проанализируйте результаты теста и сделайте выводы
if p_value < 0.05:
    print("\nГипотеза не верна")
else:
    print("\nГипотеза верна")
```
