2. Проведены испытания на каждом из четырёх уровней фактора над одной и
той же выборкой из шести элементов. Методом дисперсионного анализа при уровне
значимости 0,05 проверить основную гипотезу о равенстве групповых средних и
основную гипотезу о равенстве индивидуальных средних. Предполагается, что
результирующий признак нормально распределён в исследуемой выборке. Результаты
испытания приведены в таблице.

```python
import numpy as np
from scipy.stats import f_oneway, kruskal

# Данные
data = np.array([
    [22, 25, 19, 14, 20, 14],
    [23, 24, 20, 17, 21, 15],
    [22, 22, 22, 20, 21, 19],
    [21, 24, 25, 24, 21, 23]
])

# Групповые средние
group_means = np.array([19, 20, 21, 23])

# Индивидуальные средние
individual_means = np.array([22, 23.75, 21.5, 18.75, 20.75, 17.75])

# Проведение дисперсионного анализа для групповых средних
statistic_group, p_value_group = f_oneway(*data)

# Проведение теста Крускала-Уоллиса для индивидуальных средних
statistic_individual, p_value_individual = kruskal(*data)

# Уровень значимости
alpha = 0.05

# Вывод результатов для групповых средних
print("Групповые средние:")
print(f'Статистика : {statistic_group}')
print(f'p-значение : {p_value_group}')

# Проверка гипотезы о равенстве групповых средних
if p_value_group < alpha:
    print("Средние групп различны.")
else:
    print("Средние групп одинаковы.")

# Вывод результатов для индивидуальных средних
print("\nИндивидуальные средние:")
print(f'Статистика : {statistic_individual}')
print(f'p-значение : {p_value_individual}')

# Проверка гипотезы о равенстве индивидуальных средних
if p_value_individual < alpha:
    print("Индивидуальные средние различны.")
else:
    print("Индивидуальные средние одинаковы.")
```
